```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Read a csv file
# converting date object fields to date type
flights = pd.read_csv("ny-flights.csv.gz",
                      parse_dates=["fl_date", "arr", "dep"])
flights
```

```
flights['unique_carrier'].value_counts()
```

```
# countind coulumn data by group
flights['dep'].value_counts()
```

```
# sampling time series data by Hours like grouping data by hour and counting
flights['dep'].value_counts().resample('H').sum()
```

```
#Calculate Rolling Mean
#the first 8 values, average them, 
# then drop the 7 values and add the 9th value, etc

flights['dep'].value_counts().resample('H').sum().rolling(8).mean()
```

```
#plotting time series data of departure with rolloing avg

(flights['dep']
    .value_counts()
    .resample('H')
    .sum()
    .rolling(8).mean()
    .plot(figsize=(12, 6),
          title="Number of Flights (8H Rolling Mean)"))
sns.despine()
```
