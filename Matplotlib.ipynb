```
# random array

x = np.random.random(10)
x
```

```
n = np.random.normal(size=100) # creating univariate distribution using numpy for plot
n                              
```

```
sns.distplot(n)
```

```
# plotting uniform distribution

x = np.random.rand(100)
sns.distplot(x)
```

```
# plotting normal distribution

x = np.random.randn(1000)
sns.distplot(x)
```

```
# ploting poisson distribution

x = np.random.poisson(5, 1000)
sns.distplot(x)
```
